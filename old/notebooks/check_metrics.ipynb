```python
import pandas as pd
import glob
paths = [p for p in glob.glob('results/**/*.json') if 'dataset-metadata' not in p]
g_paths = [p for p in paths if 'metrics_g' in p]
c_paths = [p for p in paths if 'metrics_c' in p]
v_paths = [p for p in paths if 'metrics_v' in p]
g_results = pd.DataFrame({p.split('/')[1]:pd.read_json(p, typ='series',) for p in g_paths}).T
c_results = pd.DataFrame({p.split('/')[1]:pd.read_json(p, typ='series',) for p in c_paths}).T
v_results = pd.DataFrame({p.split('/')[1]:pd.read_json(p, typ='series',) for p in v_paths}).T
```
